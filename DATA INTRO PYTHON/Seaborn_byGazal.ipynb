## Seaborn Line plot

```python

# Sample data (replace this with your own DataFrame)
data = {
    'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Temperature': [25, 26, 27, 28, 29]
}

# Create a DataFrame
dfsample = pd.DataFrame(data)

# Create a line plot using Seaborn
sns.lineplot(data=dfsample, x='Day', y='Temperature', marker='o', color='blue')

# Add title and labels
plt.title("Temperature Variation by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Temperature (°C)")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

## Outlier detection and removal using boxplot



```python

# Create a synthetic dataset with outliers
np.random.seed(42)
dataoutlier = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value': np.random.normal(loc=0, scale=1, size=100)
})
dataoutlier.loc[0:5, 'Value'] = 10  # Add outliers

# Create a box plot using Seaborn
sns.boxplot(data=dataoutlier, x='Category', y='Value')

# Add title and labels
plt.title("Box Plot with Outliers")
plt.xlabel("Category")
plt.ylabel("Value")

# Show the plot
plt.show()
```

```python
# Generate synthetic data with outliers
np.random.seed(42)
data = np.concatenate([np.random.normal(loc=0, scale=1, size=80),
                       np.random.normal(loc=10, scale=1, size=20)])  # Add outliers

# Create a box plot to visualize the data and identify outliers
sns.boxplot(data=data)
plt.title("Box Plot of Data with Outliers")
plt.xlabel("Data Points")
plt.ylabel("Values")
plt.show()

# Ask the user to specify a threshold for outlier removal
threshold = float(input("Enter the threshold for outlier removal: "))

# Detect outliers based on the user-defined threshold
outliers = data[(data < np.percentile(data, 25) - 1.5 * (np.percentile(data, 75) - np.percentile(data, 25))) |
                (data > np.percentile(data, 75) + 1.5 * (np.percentile(data, 75) - np.percentile(data, 25)))]

# Remove outliers based on the user-defined threshold
data_cleaned = data[(data >= np.percentile(data, 25) - threshold) & 
                    (data <= np.percentile(data, 75) + threshold)]

# Create a box plot to visualize the cleaned data without outliers
sns.boxplot(data=data_cleaned)
plt.title("Box Plot of Cleaned Data without Outliers")
plt.xlabel("Data Points")
plt.ylabel("Values")
plt.show()

# Display the removed outliers
print("Outliers removed from the data:")
print(outliers)
```
