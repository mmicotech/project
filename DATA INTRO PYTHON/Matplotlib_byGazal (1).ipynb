# Set limits and tick labels

```python
import matplotlib.pyplot as plt

# Create some sample data
x = range(10)
y = [i**2 for i in x]

# Plot the data
plt.plot(x, y)

# Set the limits for the x-axis and y-axis
plt.xlim(2, 7)  # Limit x-axis from 2 to 7
plt.ylim(0, 50)  # Limit y-axis from 0 to 50

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create some sample data
x = range(10)
y = [i**2 for i in x]

# Plot the data
plt.plot(x, y)

# Set custom ticks for the x-axis and y-axis
plt.xticks([0, 2, 4, 6, 8, 10], ['zero', 'two', 'four', 'six', 'eight', 'ten'])  # Custom labels for x-axis
plt.yticks([0, 25, 50, 75, 100], ['0', '25', '50', '75', '100'])  # Custom labels for y-axis

plt.show()
```

```python
# Create a new figure
fig = plt.figure()

# Add axes to the figure
ax1 = fig.add_subplot(1, 1, 1)

# Plot tenure vs monthly_charges
scatter = ax1.scatter(df['tenure'], df['MonthlyCharges'], color='b')
#scatter = ax1.plot(df['tenure'], df['MonthlyCharges'], color='b')

# Set limits and tick labels
ax1.set_xlim(0, 75)  # Setting limits for tenure from 0 to 75 months
ax1.set_ylim(0, 150)  # Setting limits for monthly_charges from 0 to 150 dollars
ax1.set_xticks(range(0, 70, 10))  # Setting custom tick labels for tenure
ax1.set_yticks(range(0, 150, 10))  # Setting custom tick labels for monthly_charges

# Set figure properties
fig.suptitle('Telecom Churn Dataset: Tenure vs Monthly Charges', fontsize=16)
fig.set_size_inches(8, 6)
ax1.set_xlabel('Tenure (months)')
ax1.set_ylabel('Monthly Charges ($)')

# Display the figure
plt.grid(True)  # Add grid lines
plt.show()
```

The scatter plot visualizes the relationship between 'tenure' (the number of months a customer has stayed with the telecom service) and 'monthly_charges' (the monthly charges for the telecom service) from the telecom churn dataset.

Observations:

Tenure vs Monthly Charges: The plot shows that as 'tenure' increases, 'monthly_charges' also tend to increase. This suggests that customers who stay longer with the telecom service might be subscribing to higher-priced plans or adding more services over time.
Distribution: The scatter plot indicates that most data points are clustered towards the lower end of 'monthly_charges' and 'tenure', suggesting that many customers have lower monthly charges and shorter tenure with the service.

# Histogram

```python
# Create a new figure
fig = plt.figure()

# Add axes to the figure
ax1 = fig.add_subplot(1, 1, 1)

# Create a histogram for monthly_charges
ax1.hist(df['MonthlyCharges'], bins=10, color='g', histtype='bar')
#histtype : {'bar', 'barstacked', 'step', 'stepfilled'}

# Set figure properties
fig.suptitle('Histogram of Monthly Charges', fontsize=16)
fig.set_size_inches(8, 6)
ax1.set_xlabel('Monthly Charges ($)')
ax1.set_ylabel('Frequency')

# Display the figure
plt.grid(True)  # Add grid lines
plt.show()
```

# bar chart

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
# Calculate the average monthly_charges for each contract type
average_charges_by_contract = df.groupby('Contract')['MonthlyCharges'].mean().reset_index()

# Create a new figure
fig = plt.figure()

# Add axes to the figure
ax1 = fig.add_subplot(1, 1, 1)

# Create a bar graph for average monthly_charges by contract
ax1.bar(average_charges_by_contract['Contract'], average_charges_by_contract['MonthlyCharges'], color='b')

# Set figure properties
fig.suptitle('Average Monthly Charges by Contract Type', fontsize=16)
fig.set_size_inches(10, 6)
ax1.set_xlabel('Contract Type')
ax1.set_ylabel('Average Monthly Charges ($)')

# Display the figure
plt.grid(True)  # Add grid lines
plt.show()
```

# Scatter plot

```python
fig = plt.figure()

# Add axes to the figure
ax1 = fig.add_subplot(1, 1, 1)

# Create a scatter plot for tenure vs monthly_charges
scatter = ax1.scatter(df['tenure'], df['MonthlyCharges'],edgecolor ="g",marker ="<")

# Set colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Contract Type')
cbar.set_ticks([0, 1, 2])
cbar.set_ticklabels(df['Contract'].unique())

# Set figure properties
fig.suptitle('Scatter Plot: Tenure vs Monthly Charges by Contract Type', fontsize=16)
fig.set_size_inches(10, 8)
ax1.set_xlabel('Tenure (months)')
ax1.set_ylabel('Monthly Charges ($)')

# Display the figure
plt.grid(True)  # Add grid lines
plt.show()
```

This scatter plot visualizes the relationship between "Tenure" and "Monthly Charges" for customers, with different contract types indicated by color. 

# Pie Chart

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Apple', 'Banana', 'Orange', 'Grapes']
sizes = [30, 25, 20, 25]  # Sizes in percentage
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice (i.e. 'Apple')

# Plot
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
#labels contains the labels for each slice of the pie chart.
#sizes contains the sizes of each slice in percentages.
#colors contains the colors for each slice.
#explode specifies how much each slice should be "exploded" from the center of the pie.
#autopct specifies the format of the percentages displayed on each slice.
#shadow adds a shadow effect to the pie chart.
#startangle specifies the angle at which the first slice should be drawn.

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Fruit distribution')

plt.show()
```

```python
# Calculate the count of each contract type
contract_counts = df['Contract'].value_counts()

# Create a new figure
fig = plt.figure()

# Add axes to the figure
ax1 = fig.add_subplot(1, 1, 1)

# Create a pie chart for contract distribution
ax1.pie(contract_counts, labels=contract_counts.index, autopct='%1.1f%%', startangle=90, colors=['b', 'g', 'r'])

# Set figure properties
fig.suptitle('Pie Chart: Distribution of Contract Types', fontsize=16)
fig.set_size_inches(8, 8)

# Display the figure
plt.show()
```
